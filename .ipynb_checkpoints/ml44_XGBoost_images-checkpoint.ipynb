```python
import pandas as pd
import numpy as np

from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier
#from sklearn.ensemble import XGBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier

from sklearn.metrics import accuracy_score, roc_auc_score
from sklearn.metrics import r2_score

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold, StratifiedKFold
from sklearn.model_selection import train_test_split
```

```python
img = pd.read_csv("C:\Hogwarts\machine_learning\Cases\Image Segmentation\Image_Segmentation.csv")
```

```python
img.head()
```

```python
y = img["Class"]

X = img.drop("Class", axis = 1)
```

```python
le = LabelEncoder()
le_y = le.fit_transform(y) # NOT fit but fit_transform

print(le.classes_)
```

# XGBoost

```python
xgb = XGBClassifier(random_state = 2022)

print(xgb.get_params())
```

```python
kfold = StratifiedKFold(n_splits = 5,
                       shuffle = True,
                       random_state = 2022)
```

```python
params = {"n_estimators": [50, 100, 150],
         "max_depth": [1,2,3,4],
         "learning_rate":[0.01, 0.1, 0.15, 0.2, 0.3]}
```

```python
clf = GradientBoostingClassifier(random_state = 2022)
print(clf.get_params())
```

```python
gcv = GridSearchCV(xgb,
                  param_grid = params,
                  cv = kfold,
                  verbose = 3,
                  scoring = "neg_log_loss")

gcv.fit(X, le_y)
```

```python
import matplotlib.pyplot as plt
```

```python
best_model =gcv.best.estimator_
```
