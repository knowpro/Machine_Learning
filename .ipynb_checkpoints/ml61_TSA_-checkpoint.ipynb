```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
df = pd.read_csv("Datasets\BUNDESBANK-BBK01_WT5511.csv")
```

```python
df.head()
```

```python
df.plot()
plt.show()
```

```python
result = seasonal_decompose(df["Value"],
                           model = "additive",
                           period = 12)
result.plot()
```
