```python
import pandas as pd
import numpy as np
from sklearn.linear_model import Ridge, Lasso, ElasticNet
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.model_selection import KFold, GridSearchCV
```

```python
# Conrete_Data.csv
boss = pd.read_csv(r"C:\Hogwarts\machine_learning\Datasets\Boston.csv")
```

```python
boss.head()
```

```python
X = boss.drop('medv', axis = 1)
y = boss['medv']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                   random_state = 2022,
                                                   train_size = 0.7)
```

```python
elastic= ElasticNet()
elastic.fit(X_train, y_train)

y_pred = elastic.predict(X_test)
print(r2_score(y_test, y_pred))
```

```python
elastic = ElasticNet(alpha = 1.5)
elastic.fit(X_train, y_train)

y_pred = elastic.predict(X_test)
print(r2_score(y_test, y_pred))
```

```python
kfold = KFold(n_splits = 5, shuffle = True, random_state = 2022)
elastic = ElasticNet()
```

```python
params = {'alpha': np.linspace(0.001, 11, 20),
         'l1_ratio': np.linspace(0,1,5)}
gcv = GridSearchCV(elastic, param_grid = params, cv = kfold, scoring = 'r2')
gcv.fit(X, y)
print(gcv.best_params_)
print(gcv.best_score_)
```

```python
best_model = gcv.best_estimator_
print(best_model.coef_)
```
