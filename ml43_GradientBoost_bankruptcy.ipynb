```python
import pandas as pd
import numpy as np
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, roc_auc_score, r2_score
from sklearn.metrics import r2_score
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold, StratifiedKFold
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
banka = pd.read_csv("C:\Hogwarts\machine_learning\Cases\Bankruptcy\Bankruptcy.csv",
                   index_col = 0)
```

```python
banka.head()
```

```python
X = banka.drop(['D', 'YR'] , axis =1)
y = banka['D']
```

```python
y.value_counts()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    stratify = y,
                                                    random_state = 2022,
                                                    train_size = 0.7)
```

```python
clf  = GradientBoostingClassifier(random_state = 2022)

clf.fit(X_train , y_train)
```

```python
y_pred = clf.predict(X_test)
print(accuracy_score(y_test, y_pred))
```

```python
y_pred_prob = clf.predict_proba(X_test)[:,1]
print(roc_auc_score(y_test, y_pred_prob))
```

# Grid Search CV

```python
kfold = StratifiedKFold(n_splits = 5,
             random_state = 2022,
             shuffle =True)

params = {"n_estimators": [50, 100, 150],
         "max_depth": [1,2,3,4],
         "learning_rate": [0.01, 0.1, 0.15, 0.2, 0.3]}
```

```python
clf  = GradientBoostingClassifier(random_state = 2022)

print(clf.get_params())
```

```python
gcv = GridSearchCV(clf,
                   param_grid = params,
                   cv = kfold,
                   verbose = 3,
                   scoring = "r2")
```

```python
gcv.fit(X, y)
```

```python
gcv.best_params_
```

```python
gcv.best_score_
```
