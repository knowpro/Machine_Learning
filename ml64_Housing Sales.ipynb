```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"Cases\House Property Sales Time Series\raw_sales.csv",
                 parse_dates = ["datesold"])
```

```python
df
```

```python
df["year"] = df["datesold"].dt.year
df["month"] = df["datesold"].dt.month
df.head()
```

```python
total_sales = df.groupby(["year", "month"])["price"].sum()
total_sales = total_sales.reset_index()
total_sales.tail()
```

```python
# 
```

```python
y = total_sales["price"]
y_train = y[:-6]
y_test = y[-6:]
```

```python
plt.plot(y_train,
        color = "blue",
        label = "Train")

plt.plot(y_test,
        color = "orange",
        label = "Test")

plt.legend(loc = "best")
plt.show()
```
