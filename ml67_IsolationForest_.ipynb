```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.ensemble import IsolationForest
```

```python
mow = pd.read_csv(r"Datasets\RidingMowers.csv")
```

```python
mow.head()
```

```python
X = mow.drop("Response", axis = 1)
y = mow["Response"]
```

```python
clf = IsolationForest(contamination = 0.5,
                      random_state = 2022)
```

```python
clf.fit(X)
```

```python
pred_outliers = clf.predict(X)

mow["outliers"] = pred_outliers
#mow["outliers"] = mow["outliers"].astype(str)
```

```python
sns.scatterplot(x= "Income",
               y = "Lot_Size",
               hue = "outliers",
               data = mow)
```
