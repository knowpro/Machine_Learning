```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
fp_df = pd.read_csv(r"C:\Hogwarts\machine_learning\11. Association Rules\Association Datasets\Cosmetics.csv",
           index_col = 0)
```

```python
fp_df.head()
```

## Create Frequent ItemSet

```python
itemset = apriori(fp_df,
                min_support = 0.1,
                use_colnames = True)
itemset#.head()
```

# Convert into rules

```python
rules = association_rules(itemset)
rules.head()
```

```python
rules.sort_values( by = ["lift", "confidence"],
                 ascending = False)
```
