```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
fp_df = pd.read_csv(r"C:\Hogwarts\machine_learning\11. Association Rules\Association Datasets\Faceplate.csv",
           index_col = 0)
```

```python
fp_df.head()
```

```python
fp_df = fp_df.astype(bool)
```

## Create Frequent Itemset

```python
itemset = apriori(fp_df, min_support = 0.2,
                 use_colnames=True)
itemset.head()
```

## Convert into rules

```python
rules = association_rules(itemset)
rules.head()
```
