```python
import pandas as pd
import numpy as np

from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score

from sklearn.model_selection import train_test_split
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC

import os
```

```python
os.chdir(r"C:\Hogwarts\machine_learning\Cases\Bankruptcy")
bankr = pd.read_csv("Bankruptcy.csv",
                    index_col = 0)
bankr.head()
```

```python
X = bankr.drop(["D",
                "YR"],
               axis = 1)

y= bankr["D"]
```

```python
X_train,X_test, y_train, y_test = train_test_split(X,
                                                   y,
                                                   stratify=y,
                                                   train_size=0.7,
                                                   random_state=2022)
```

```python
svm = SVC(kernel = "linear",
         probability = True,
         random_state = 2022)
svm.fit(X_train, y_train)
```

## Grid Search CV

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = {"C": np.linspace(0.001, 10, 20)}
```

```python
kfold = StratifiedKFold(n_splits = 5,
                       shuffle = True,
                       random_state = 2022)
```

# Without Scaling

### 1. Using linear

```python
svm = SVC(kernel = "linear",  # by default it is set to rbf
         probability = True,
         random_state = 2022)

gcv = GridSearchCV(svm,
                  param_grid = params,
                  cv = kfold,
                  verbose = 3,
                  scoring = "roc_auc")
gcv.fit(X, y)
```

```python
print(gcv.best_params_)
```

```python
print(gcv.best_score_)
```

### 2. Using polynomial boundary

```python
svm = SVC(kernel = "poly",
         probability = True,
         random_state = 2022)

gcv = GridSearchCV(svm,
                  param_grid = params,
                  cv = kfold,
                  verbose = 3,
                  scoring = "roc_auc")
gcv.fit(X, y)
```

```python
print(gcv.best_params_)
```

```python
print(gcv.best_score_)
```

### 3. Using radial boundary function ( rbf )

```python
# need to change params to make it suitable for radial boundary f
params = {"C": np.linspace(0.001, 10, 20),
         "gamma": np.linspace(0.001, 10, 20)}
```

```python
svm = SVC(kernel = "rbf",  # by default it is set to rbf
         probability = True,
         random_state = 2022)

gcv = GridSearchCV(svm,
                  param_grid = params,
                  cv = kfold,
                  verbose = 3,
                  scoring = "roc_auc")
gcv.fit(X, y)
```

```python
print(gcv.best_params_)
```

```python
print(gcv.best_score_)
```

# With Scaling

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing  import StandardScaler
```

```python
scaler = StandardScaler()
```

## 1. Using Linear boundary

```python
svm = SVC(kernel = "linear",
         probability = True,
         random_state = 2022)
```

```python
# using Pipe
pipe = Pipeline([("STD", scaler),
                ("SVM", svm)])

print(pipe.get_params())
```

```python
params = {"SVM__C": np.linspace(0.001, 10, 20)}

kfold = StratifiedKFold(n_splits = 5,
                       shuffle =True,
                       random_state = 2022)
```

```python
gcv = GridSearchCV(pipe,
                  param_grid = params,
                  cv =kfold,
                  verbose = 3,
                  scoring ="roc_auc")

gcv.fit(X, y)
```

```python
print(gcv.best_params_)
```

```python
print(gcv.best_score_)
```

```python
params = {"C": np.linspace(0.001, 10, 20),
         "degree": [2, 3, 4],
         "coef0": np.linspace(-2, 4, 5)}
```

```python
svm = SVC(kernel = "poly",
         probability = True,
         random_state = 2022)

gcv = GridSearchCV(svm,
                  param_grid = params,
                  cv = kfold,
                  verbose = 3,
                  scoring = "roc_auc")
gcv.fit(X, y)
```

```python
print(gcv.best_params_)
```

```python
print(gcv.best_score_)
```
